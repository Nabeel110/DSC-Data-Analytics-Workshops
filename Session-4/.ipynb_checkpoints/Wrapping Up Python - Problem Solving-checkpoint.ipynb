## Going Back to our First Problem we solved in First Session:
- Random Guessing Number Game
### Problem Statement: 
- Take a user input and ask him to guess a number between 1 and 100. If he/she guesses correctly the person wins. If not ask the user to reinput number. In case the user guess is incorrect, hint him by telling wether his/her guessed number is too high or too low, so that in the next guess he may adjust accordingly.

```python
# Solution
import random

random_number = random.randint(1,101)


while True:
    user_input = int(input('Enter a number between  1 and 100: '))
    
    if user_input == random_number:
        print('You have guessed the correct number ' + str(random_number))
        break
    elif user_input > random_number:
        print('Too High!. Guess a lower number')
    else:
        print('Too Low!. Guess a Higher number')




```

# Improvising above Program to quit the game if user inputs 'q'

```python
import random
magic_num = random.randint(1,101)

print('\t\t\t\t\t Guessing Game\t\t')
print('\t\t\t\t   By Nabeel Ahmed')

user_input = ''

while user_input!='q':
    user_input = input('Enter a number in 1 to 100 inclusive or press q to exit: ')
    if user_input == 'q':
        break
    elif int(user_input) > magic_num:
        print('Too High! Chose smaller number.')
        continue
    elif int(user_input) < magic_num:
        print('Too Low! Chose Larger number.')
        continue
    
    else:
        print('Congratulations! You have guessed the number {} correctly!'.format(magic_num))
        break
```

## Problem # 2

```python
"""
3.Question 3
Complete the function digits(n) that returns how many digits the number has. 
For example: 25 has 2 digits and 144 has 3 digits. Tip: you can figure out the digits of a number by dividing 
it by 10 once per digit until there are no digits left.

"""
# Solution:
# def digits(number):
#     str_num = str(number)
#     return len(str_num)
    











#sol 1
def digits(number):
    count =0
    if number == 0:
        return 1
    while number > 0:
        count +=1
        number = number//10
    return count
        
    
# Sol 2
# def digits(n):
#     num_string = str(number)
#     return len(num_string)

print(digits(5000))
      
```

```python

def joining_filename():
    new_filenames = []
    filenames = ["program.c", "stdio.hpp", "sample.hpp", "a.out", "math.hpp", "hpp.out",]
    for oldfilename in filenames:
        if '.hpp' in oldfilename:
            new_filenames.append((oldfilename, oldfilename.replace('.hpp','.h')))
        else:
            new_filenames.append((oldfilename,oldfilename))
            
    return new_filenames
            
        
    
    
files = joining_filename()

print(files)


```

```python
#
email_list = {"gmail.com": ["clark.kent", "diana.prince", "peter.parker"], "yahoo.com": ["barbara.gordon", "jean.grey"], "hotmail.com": ["bruce.wayne"]}
new_list = []
for key, value in email_list.items():
    for name in value:
        temp = name + "@" +key
        new_list.append(temp)
        temp =""
        
    
    
new_list



```

```python

# Solution


def email_list(domains):
    emails = []
    for domain, users in domains.items():
        for name in users:
            emails.append(name + "@"+domain)
        
    return emails
    
```

```python
print(email_list({"gmail.com": ["clark.kent", "diana.prince", "peter.parker"], "yahoo.com": ["barbara.gordon", "jean.grey"], "hotmail.com": ["bruce.wayne"]}))
```
